```python
import numpy as np
import datetime
import dateutil.parser

def getTimeofDayDistribution(purpose):
    
    purposeSegments = list(segments.find({'mainPurpose': purpose}))
    frequency = np.zeros((24))
    
    for segment in purposeSegments:
        startTime = dateutil.parser.parse(segment['startTime'])
        endTime = dateutil.parser.parse(segment['endTime'])
        while startTime < endTime:
            frequency[startTime.hour] += 1
            startTime = startTime + datetime.timedelta(0, 0, 0, 0, 0, 1, 0)
    return frequency
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nPurposes = len(purposes)
counter = 0
plt.figure(num = 1, figsize = (16, 5))  
for purpose in purposes: 
    frequency = getTimeofDayDistribution(purpose)
    plt.subplot(1, nPurposes, counter)        
    plt.bar(np.arange(0, 24, 1), frequency, color = 'k')
    plt.xlim(0, 24)
    #plt.ylim(0, 60)
    plt.xticks(np.arange(0, 25, 6), fontsize = 10)
    plt.yticks(fontsize = 10)
    plt.title(purpose, fontsize = 10)
    plt.xlabel('Time of day', fontsize = 10) 
    counter += 1
```

```python
def getDayOfWeekDistribution(purpose):
    
    purposeSegments = list(segments.find({'mainPurpose': purpose}))
    frequency = np.zeros((7))
    weekDays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    
    for segment in purposeSegments:
        startTime = dateutil.parser.parse(segment['startTime'])
        endTime = dateutil.parser.parse(segment['endTime'])
        while startTime < endTime:
            currentDay = startTime.strftime('%A')
            frequency[weekDays.index(currentDay)] += 1
            startTime = startTime + datetime.timedelta(1)
    return frequency
```

```python
nPurposes = len(purposes)
counter = 0
plt.figure(num = 1, figsize = (16, 5))  
for purpose in purposes: 
    frequency = getDayOfWeekDistribution(purpose)
    plt.subplot(1, nPurposes, counter)        
    plt.bar(np.arange(0, 7, 1), frequency, color = 'k')
    #plt.xlim(0, 24)
    #plt.ylim(0, 60)
    #plt.xticks(np.arange(0, 25, 6), fontsize = 10)
    plt.yticks(fontsize = 10)
    plt.title(purpose, fontsize = 10)
    plt.xlabel('Day of week', fontsize = 10) 
    counter += 1
```

```python
def getDurationDistribution(phNum):
    
    activities = list(segments.find({'type': 'Activity', 'phNum': phNum}))
    nearbyDurations, totalDuration = [], 0
    
    for activity in activities:
        totalDuration += activity['duration']
        nearbyDuration = 0
        centroid = activity['centroid']['coordinates']
        query = {'centroid': {'$nearSphere': {'$geometry': {'type': 'Point',
                                                            'coordinates': [centroid[0], centroid[1]]},
                                              '$maxDistance': 100}},
                 'phNum': phNum}
        nearbyActivities = segments.find(query)
        for nearbyActivity in nearbyActivities:
            nearbyDuration += nearbyActivity['duration']
        nearbyDurations.append(nearbyDuration)
    
    for i in range(0, len(nearbyDurations)):
        nearbyDurations[i] = (100 * nearbyDurations[i]) / totalDuration
        
    duration = {}
    for purpose in purposes:
        duration[purpose] = []

    for i in range(0, len(activities)):
        if 'mainPurpose' in activities[i]:
            duration[activities[i]['mainPurpose']].append(nearbyDurations[i])
    
    return duration
```

```python
def plotDurationHistogram(duration):
    nPurposes = len(purposes)
    counter = 0
    plt.figure(num = 1, figsize = (16, 5))  
    for purpose in purposes: 
        plt.subplot(1, nPurposes, counter)
        tPurpose = np.zeros(len(duration[purpose]))
        for i in range(0, len(duration[purpose])):
            tPurpose[i] = duration[purpose][i]
        if len(duration[purpose]) > 0:
            plt.hist(tPurpose, normed = 1, color = 'k')
        plt.xlim(0, 100)
        #plt.ylim(0, 60)
        plt.xticks(np.arange(0, 101, 25), fontsize = 10)
        plt.yticks(fontsize = 10)
        plt.title(purpose, fontsize = 10)
        plt.xlabel('%age of total time', fontsize = 10) 
        counter += 1
```

```python
durationTotal = {}
for purpose in purposes:
    durationTotal[purpose] = []

phNums = segments.distinct('phNum')
for phNum in phNums:
    durationPhNum = getDurationDistribution(phNum)
    for purpose in purposes:
        durationTotal[purpose] += durationPhNum[purpose]
```

```python
plotDurationHistogram(durationTotal)
```
